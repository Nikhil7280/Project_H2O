```python
import numpy as np
import pandas as pd
from random import randint, shuffle
```

```python
feature_1 = [1]*30
feature_2_18 = [3]*30
feature_2_19 = [4]*30
feature_3 = [14]*30
feature_4 = ['yes']*30
feature_5 = ['rainy']*30
feature_6 = []
feature_7 = []
feature_8 = ['urban']*30
feature_9 = []

feature_y = []
    
```

```python
weather = ['hot', 'cloudy', 'rain', 'humid', 'dry', 'cold']
for i in range(30):
    feature_7.append(randint(5000, 7000))
    feature_y.append(randint(2000, 3000))
```

```python
for i in range(10):
    feature_6.append('rain')
for i in range(8):
    feature_6.append('cloudy')
for i in range(6):
    feature_6.append('humid')
for i in range(2):
    feature_6.append('cold')
    feature_6.append('hot')
    feature_6.append('dry')
```

```python
shuffle(feature_6)
```

```python
for i in range(21):
    feature_9.append('no')
for i in range(9):
    feature_9.append('yes')
shuffle(feature_9)
```

```python
features = [feature_1, feature_2_18, feature_2_19, feature_3, feature_4, feature_5, feature_6, feature_7, feature_8, feature_9, feature_y]
```

```python
features = np.array(features)
```

```python
X = pd.DataFrame(features.transpose(), columns = ['ID',"No of Members Under 18", "No of Members above 18","No of Water using Appliances","Garden(Yes/No)","Season","Weather","Water Availability in Diff Sources","Region","Holiday(Yes/No)","Per Day Water Usage"])
```
