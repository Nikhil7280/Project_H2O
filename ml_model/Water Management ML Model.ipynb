## Importing the Necessary Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Importing the Dataset

```python
dataset = pd.read_csv("Water Usage Data for ML Model")
```

```python
dataset = dataset.drop('Unnamed: 0', axis=1)
```

```python
dataset
```

## Encoding Feature Variables

```python
X = dataset.iloc[:, 1:-1]
y = dataset.iloc[:, -1]
```

```python
categorical = ["Garden(Yes/No)", "Season", "Weather", "Region", "Holiday(Yes/No)"]
```

```python
from sklearn.preprocessing import LabelEncoder

le_season = LabelEncoder()
X["Season"] = le_season.fit_transform(X["Season"])

le_weather = LabelEncoder()
X["Weather"] = le_weather.fit_transform(X["Weather"])

le_region = LabelEncoder()
X["Region"] = le_region.fit_transform(X["Region"])

le_garden = LabelEncoder()
X["Garden(Yes/No)"] = le_garden.fit_transform(X["Garden(Yes/No)"])

le_holiday = LabelEncoder()
X["Holiday(Yes/No)"] = le_holiday.fit_transform(X["Holiday(Yes/No)"])
```

```python
X
```

## Training the Model

```python
from sklearn.linear_model import LinearRegression

regressor = LinearRegression()
regressor.fit(X,y)
```

```python
with open('model_pickle.pkl', 'wb') as f:
    pickle.dump(regressor,f)
```
